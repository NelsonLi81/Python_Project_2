```python
import pandas as pd
import matplotlib.pyplot as plt

# Question 1
# Load the CSV file into a DataFrame
dataset = pd.read_csv('C:\\Users\\NelsonCN\\Desktop\\CanTek_DataAnalaysis\\Week 3 - Python\\Assignement 1\\Question_1_TotRev_3.csv')

# Convert the 'payment_day' column to datetime format
dataset['payment_day'] = pd.to_datetime(dataset['payment_day'])

# Pivot the data to create a DataFrame with one column per country
plot_data = dataset.pivot(index='payment_day', columns='country', values='total_revenue')

# Plot the time-series data
plot_data.plot(kind='line', figsize=(10, 6))
plt.title('Daily Total Revenue for Top 3 Countries')
plt.xlabel('Payment Day')
plt.ylabel('Total Revenue')
plt.legend(title='Country')
plt.show()
```

```python
import pandas as pd

# Define a set of positive and negative keywords
positive_keywords = ['amazing', 'awesome', 'best', 'excellent', 'fantastic', 'great', 'happy', 'love', 'perfect', 'wonderful']
negative_keywords = ['awful', 'bad', 'boring', 'disappointing', 'horrible', 'terrible', 'sad', 'hate', 'worst']

# Define a function to classify the sentiment of a piece of text
def classify_sentiment(text):
    # Convert the text to lowercase
    text = text.lower()
    
    # Count the number of positive and negative keywords in the text
    positive_count = sum(text.count(keyword) for keyword in positive_keywords)
    negative_count = sum(text.count(keyword) for keyword in negative_keywords)
    
    # Classify the sentiment based on the counts
    if positive_count > negative_count:
        return 'Positive'
    elif negative_count > positive_count:
        return 'Negative'
    else:
        return 'Neutral'

# Load the movie data into a DataFrame
movies = pd.read_csv('C:\\Users\\NelsonCN\\Desktop\\CanTek_DataAnalaysis\\Week 3 - Python\\Assignement 1\\Question_2_Desc.csv')

# Add a new column to the DataFrame containing the sentiment classification
movies['sentiment'] = movies['description'].apply(classify_sentiment)

# Print out the resulting DataFrame
print(movies)
```

```python
import pandas as pd
import matplotlib.pyplot as plt
import matplotlib.cm as cm

# Question 3: assumption to display the total Revenue>= 1000 only
# Load the dataset from the CSV file
df = pd.read_csv('C:\\Users\\NelsonCN\\Desktop\\CanTek_DataAnalaysis\\Week 3 - Python\\Assignement 1\\Question 3\\DataSet_3_RevCoun.csv')  # Replace 'your_dataset.csv' with the actual file path

# Group the data by country and calculate total revenue for each country
country_revenue = df.groupby('country')['total_revenue'].sum().sort_values()

# Create a colormap ranging from light green to dark green
cmap = cm.get_cmap('Greens', len(country_revenue))

# Create a bar plot with gradient-colored bars
plt.figure(figsize=(12, 6))
ax = country_revenue.plot(kind='bar', color=[cmap(i) for i in range(len(country_revenue))])
plt.title('Total Revenue from Different Countries')
plt.xlabel('Country')
plt.ylabel('Total Revenue')
plt.xticks(rotation=45)
plt.tight_layout()

# Add revenue values on top of the bars
for p in ax.patches:
    ax.annotate(f'${p.get_height():.2f}', (p.get_x() + p.get_width() / 2., p.get_height()), ha='center', va='center', xytext=(0, 10), textcoords='offset points')

plt.show()
```
